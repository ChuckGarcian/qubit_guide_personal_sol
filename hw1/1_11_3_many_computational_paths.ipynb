```python
import matplotlib.pyplot as plt
import numpy as np
import math

def get_y_axis (psi, n):
  numerator = pow(np.sin ((n*psi) / 2), 2)
  denomator = pow(np.sin (psi / 2), 2)
  
  return (numerator/denomator) / pow(n, 2)
plt.figure ()

x_values = np.linspace (0, 2*np.pi, 100)
number_n = 10 # Number of n's to graph

for n in range (1, number_n):
  y_values = get_y_axis (x_values, n)
  
  
  plt.plot (x_values, y_values, label="n={}".format(n))

plt.title (r"Probability as a function of angle psi accross different $n\in [1, 10]$")
plt.ylabel(r'$P(\psi)=\frac{1}{n^2}\frac{\sin^2\left(\frac{n\psi}{2}\right)}{\sin^2\left(\frac{\psi}{2}\right)}$', usetex=True)
plt.xlabel (r'Angle: $\psi$')
plt.axvline(x=np.pi, color='red', linestyle='--', label=r'$x=\pi$')
plt.legend(loc='lower right')
plt.show ()
```
